```python
s = input("Enter a string: ")
vowels = "aeiouAEIOU"
v = 0
c = 0
for ch in s:
    if ch.isalpha():
        if ch in vowels:
            v += 1
        else:
            c += 1
print("Vowels:", v)
print("Consonants:", c)
```

```python
scores = {"Alice": 85, "Bob": 92, "Charlie": 78}
max_score = max(scores.values())
top_student = max(scores, key=scores.get)
print(f"The maximum value is {max_score} (achieved by {top_student}).")
```
