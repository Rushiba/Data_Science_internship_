```python
class LoginFailedError(Exception):
    pass

correct_password = "admin"
attempts = 3

while attempts > 0:
    try:
        pwd = input(f"Enter password ({attempts} attempts left): ")
        if pwd == correct_password:
            print("Login successful!")
            break
        else:
            attempts -= 1
            if attempts == 0:
                raise LoginFailedError("Account locked. Too many failed attempts.")
            print("Incorrect password.")
    except LoginFailedError as e:
        print(f"Security Alert: {e}")
```

```python
def convert_to_float(data):
    try:
        return float(data)
    except (ValueError, TypeError):
        return "Error: Conversion failed. Input is not a valid float string."

print(convert_to_float("12.5"))
print(convert_to_float("abc"))
```

```python
class InsufficientBalanceError(Exception):
    pass

balance = 1000
try:
    amount = float(input("Enter withdrawal amount: "))
    if amount > balance:
        raise InsufficientBalanceError(f"Cannot withdraw Rs{amount}. Balance is only ${balance}.")
    balance -= amount
    print(f"Withdrawal successful. Remaining balance: Rs{balance}")
except InsufficientBalanceError as e:
    print(e)
```

```python
class InvalidAgeError(Exception):
    pass
try:
    age = int(input("Enter your age: "))
    if age < 0 or age > 120:
        raise InvalidAgeError("Age must be between 0 and 120.")
    print(f"Age {age} recorded.")
except InvalidAgeError as e:
    print(f"Validation Error: {e}")
except ValueError:
    print("Error: Age must be a number.")
```

```python

student_data = "Anu,85\nRushiba,92\nswany,78"

with open("marks.txt", "w") as file:
    file.write(student_data)

total = 0
count = 0
with open("marks.txt", "r") as file:
    for line in file:
        name, mark = line.strip().split(",")
        total += int(mark)
        count += 1

print(f"Average Mark: {total / count}")
```

```python
import os

users = {}

while True:
    print("\n1.Register  2.Login  3.Exit")
    ch = input("Enter choice: ")

    if ch == "1":
        u = input("Username: ")
        if u in users:
            print("User already exists")
        else:
            p = input("Password: ")
            users[u] = p
            with open("users.txt", "a") as f:
                f.write(u + "," + p + "\n")
            print("Registration Successful")

    elif ch == "2":
        for i in range(3):
            u = input("Username: ")
            p = input("Password: ")
            if u in users and users[u] == p:
                print("Login Successful")
                break
            else:
                print("Wrong Credentials")
        else:
            print("Login Failed")

    elif ch == "3":
        print("Thank You")
        break

    else:
        print("Invalid Choice")
```
